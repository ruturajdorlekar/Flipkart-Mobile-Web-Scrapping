### Use the requests library to download web pages

```python
#importing the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
page_num=int(input('Enter the number of pages: '))
mobile_names=[]
mobile_specifications=[]
mobile_prices=[]
mobile_ratings=[]
mobile_votes_and_reviews=[]

for i in range(1, page_num+1):
    url = 'https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3Drealme&otracker=nmenu_sub_Electronics_0_Realme&page'+str(i)
    response= requests.get(url)
    page_contents=response.text                   #the content written in html
    doc = BeautifulSoup(page_contents,'html.parser')
    mobile_name= doc.find_all('div',{'class':'_4rR01T'})
    mobile_specification= doc.find_all('ul',{'_1xgFaf'})
    mobile_price= doc.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    mobile_rating= doc.find_all('div',{'_3LWZlK'})
    mobile_votes_and_review= doc.find_all('span',{'_2_R_DZ'})
    print('Phones in page {}: {}'.format(i, len(mobile_name)))
    
    print()
    for i in mobile_name:
        mobile_names.append(i.text)
    print(mobile_names[:10])
        
    print()
    for i in mobile_specification:
        mobile_specifications.append(i.text)
    print(mobile_specifications[:10])
        
    print()
    for i in mobile_price:
        mobile_prices.append(i.text)
    print(mobile_prices[:15])
        
    print()
    for i in mobile_rating:
        mobile_ratings.append(i.text)
    print(mobile_ratings[:15])
        
    print()
    for review in mobile_votes_and_review:
        mobile_votes_and_reviews.append(review.text)
    print(mobile_votes_and_reviews[:15])
```

```python
df1 = {'Mobile_name':mobile_names,
          'Price':mobile_prices,
          'Rating':mobile_ratings,
          'Specifications':mobile_specifications,
          'Reviews':mobile_votes_and_reviews
         }
```

```python
df= pd.DataFrame(df1)
```

```python
df.head()
```

```python
Mobile_Name=[]
Mobile_Colour=[]

for name in df['Mobile_name']:
    Name=name.split('(')
    Mobile_name=Name[0]
    Mobile_Name.append(Mobile_name)
print(Mobile_Name[:15])

print() 
for color in df['Mobile_name']:
    C=color.split('(')[1]
    Colour=C.split(',')
    Mobile_colour=Colour[0]
    Mobile_Colour.append(Mobile_colour)
print(Mobile_Colour[:15])
```

```python
STORAGE_RAM=[]
STORAGE_ROM=[]
DISPLAY_INCH=[]
DISPLAY_QUALITY=[]
FRONT_CAMERA=[]
REAR_CAMERA=[]
BATTERY=[]
PROCESSOR=[]
WARRANTY=[]
RATING_COUNT=[]

for storage in df['Specifications']:
    Storage = storage.split('RAM')[0][0:-1]
    STORAGE_RAM.append(Storage)
print(STORAGE_RAM[:15])
        
print()
for storage_rom in df['Specifications']:
    Storage = storage_rom.split('ROM')[0][-6:-1]
    STORAGE_ROM.append(Storage)
print(STORAGE_ROM[:15])
    
print()
for display in df['Specifications']:
    Display = display.split('(')[1]
    Display_inch=Display.split('inch')[0][:-1]
    DISPLAY_INCH.append(Display_inch)
print(DISPLAY_INCH[:15])
    
print()
for quality in df['Specifications']:
    Quality = quality.split(')')[1]
    Display_quality=Quality.split('+')[0]  
    DISPLAY_QUALITY.append(Display_quality)
print(DISPLAY_QUALITY[:10])

print()
for front in df['Specifications']:
    Front = front.split('Front')[0][-4:-1]
    FRONT_CAMERA.append(Front)
print(FRONT_CAMERA[:10])

print()
for camera in df['Specifications']:
    Camera = camera.split('Display')[1]
    C=Camera.split('|')[0]
    R=C.split('Rear')[0][:-1]
    REAR_CAMERA.append(R)
print(REAR_CAMERA[:50])
    
print()
for battery in df['Specifications']:
    Battery = battery.split('Front')[1][7:15]
    BATTERY.append(Battery)
print(BATTERY[:10])
    
print()
for proc in df['Specifications']:
    Processor = proc.split('Battery')[1]
    p=Processor.split('Processor')[0][:-1]
    PROCESSOR.append(p)
print(PROCESSOR[:10])
    
print()
for war in df['Specifications']:
    warrenty = war.split('Processor')[1][:6]
    WARRANTY.append(warrenty)
print(WARRANTY[:10])

print()
for Votes in df['Reviews']:
    votes = Votes.split('Ratings')[0][:-1]
    RATING_COUNT.append(votes)
print(RATING_COUNT[:10])
```

```python
df2 = {'MOBILE_NAME':Mobile_Name,
          'COLOUR':Mobile_Colour,
          'PRICE':mobile_prices,
          'RATING':mobile_ratings,
          'RATING_COUNT':RATING_COUNT,
          'RAM':STORAGE_RAM,
          'ROM':STORAGE_ROM,
          'DISPLAY':DISPLAY_INCH,
          'DISPLAY_QLTY':DISPLAY_QUALITY,
          'FRONT_CAMERA':FRONT_CAMERA,
          'REAR_CAMERA':REAR_CAMERA,
          'BATTERY':BATTERY,
          'PROCESSOR':PROCESSOR,
          'WARRANTY':WARRANTY
         }
```

```python
# using dictionary to convert specific columns          or changing Data Type
convert_dict = {
                        'RATING': float,
                        'DISPLAY': float
                       }
 
realme = realme.astype(convert_dict)
```

```python
realme["RAM"] = realme['RAM'].str.replace('GB', '').astype('int')
```

```python
realme["ROM"] = realme['ROM'].str.replace('GB', '').astype('int')
```

```python
realme["BATTERY"] = realme['BATTERY'].str.replace('mAh', '').astype('int')
```

```python
realme["WARRANTY"] = realme['WARRANTY'].str.replace('Year', '').astype('int')
```
